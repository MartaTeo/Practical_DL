## `y_target` diagnostics

**Distribution**
- The target is extremely leptokurtic: the vast majority of observations cluster tightly near zero, but the tails extend to roughly ±2,300.
- Even clipped at ±50, the distribution looks like a sharp spike, confirming that typical returns are tiny and outliers are massive relative to the median.
- The positive tail on a log-y scale decays slowly, consistent with a heavy-tailed or power-law distribution rather than a Gaussian.

**By horizon**
- Median `y_target` is negative at all horizons, ranging from roughly -0.0002 (horizon 1) to -0.0018 (horizon 25).
- Longer horizons have more negative medians, suggesting a structural downward drift that grows with holding period.
- This is an important modeling constraint: horizon should not be treated as a generic categorical but as an ordered variable with a systematic effect.

**Over time (ts_index)**
- The median `y_target` is broadly stable near zero but shows intermittent sharp spikes up to ±15, concentrated at specific time points.
- These spikes are not evenly distributed and likely correspond to market stress events or structural breaks.
- Models that don't account for time-varying noise or regime changes will be adversely affected by these periods.

**By sub_category**
- Median targets are all slightly negative, ranging from about -0.003 (PHHHVYZI) to near zero (V8BKY1IV).
- The spread across categories is small in absolute terms but meaningful relative to the overall median.
- `sub_category` appears to carry a small but persistent conditional signal.

**Quantiles over time**
- q50 (orange) is nearly flat at zero throughout the full time range.
- q10 and q90 are broadly symmetric and stable, but show extreme spikes at a few time points, especially near ts_index ~500, ~2000, and ~3500.
- These spike events dominate the tails and will strongly influence any loss function that is sensitive to large residuals (e.g. MSE).

**Modeling implications**
- MSE loss will be dominated by tail events; consider Huber loss, quantile regression, or explicit outlier treatment.
- Horizon must be used as a feature with monotonic regularization or at minimum as a separate stratum.
- Spike time periods should be investigated and potentially down-weighted via `weight` or excluded from validation.

## `y_target` distribution by horizon

**Shape is identical across all horizons**
- At every horizon (1, 3, 10, 25), the clipped distribution shows the exact same sharp spike at zero with
  symmetric shoulders — the shape of the distribution does not change with horizon, only its scale does.
- On the log-y full-range plots, all four horizons show the same symmetric tent/Laplace-like shape,
  confirming the heavy-tail structure is consistent and not horizon-specific.

**Scale grows with horizon**
- The x-axis range widens monotonically: horizon=1 extends to roughly ±1000, horizon=3 to ±1250,
  horizon=10 to ±2000, and horizon=25 to ±2500.
- This is consistent with cumulative returns: longer horizons accumulate more variance, so extreme values
  become larger in absolute terms.
- The core of the distribution (the spike near zero) becomes slightly wider at longer horizons, but the
  proportional shape is preserved.

**No pathological differences between horizons**
- There are no bimodal structures, no horizon-specific outlier clusters, and no sign of data quality issues
  specific to one horizon.
- The clipped plots show small bumps at the ±10 boundary for all horizons, which are artefacts of the
  clipping itself, not real data features.

**Modeling implications**
- Because shape is preserved but scale grows, **target normalization within horizon is the correct approach**
  before pooled multi-horizon training — e.g. divide `y_target` by its per-horizon standard deviation or
  apply a rank normalization within each horizon.
- A model trained on raw pooled targets without normalization will implicitly over-weight long-horizon rows
  since their residuals are larger by construction.
- The Laplace-like shape (heavier tails than Gaussian, sharper peak) suggests **MAE or Huber loss will
  outperform MSE** as a training objective, since MSE will be dominated by the extreme tail events.

```python
import pandas as pd
from scipy.stats import spearmanr

num_features = df.select_dtypes(include="number").columns.drop(
    ["y_target", "weight", "horizon", "ts_index"], errors="ignore"
).tolist()

horizons = sorted(df["horizon"].unique())
ic_rows = []

for h in horizons:
    sub = df[df["horizon"] == h][num_features + ["y_target"]].dropna()
    row = {"horizon": h}
    for col in num_features:
        r, _ = spearmanr(sub[col], sub["y_target"])
        row[col] = r
    ic_rows.append(row)

ic_df = pd.DataFrame(ic_rows).set_index("horizon").T
ic_df["mean_abs_IC"] = ic_df.abs().mean(axis=1)
ic_df = ic_df.sort_values("mean_abs_IC", ascending=False)

print("Top 20 features by mean absolute IC across horizons:")
display(ic_df.head(20))

# Heatmap of IC
fig, ax = plt.subplots(figsize=(10, 14))
import matplotlib.cm as cm
im = ax.imshow(ic_df.drop(columns="mean_abs_IC").head(30).values,
               aspect="auto", cmap="RdBu_r", vmin=-0.1, vmax=0.1)
ax.set_xticks(range(len(horizons)))
ax.set_xticklabels([f"h={h}" for h in horizons])
ax.set_yticks(range(30))
ax.set_yticklabels(ic_df.head(30).index)
plt.colorbar(im, ax=ax)
ax.set_title("Spearman IC: top 30 features × horizon")
plt.tight_layout()
plt.show()
```

## Spearman IC heatmap: features × horizon

Each cell shows the **Spearman rank correlation** between one feature and `y_target`, computed
separately per horizon. Red = positive predictive relationship, blue = negative, white = no signal.
The scale is ±0.10, so all signals are weak in absolute terms but consistent across 5M+ rows.

### Group 1 — consistently positive (red across all horizons)
`feature_bz`, `feature_ca`, `feature_by`, `feature_cc`, `feature_cd`, `feature_az`, `feature_cb`,
`feature_bq`, `feature_br`
- Signal is stable regardless of horizon → robustly useful features, include in any model.

### Group 2 — consistently negative (blue across all horizons)
`feature_am`, `feature_u`, `feature_af`, `feature_ag`, `feature_bl`, `feature_an`, `feature_bn`,
`feature_t`, `feature_bm`, `feature_q`, `feature_bo`, `feature_v`, `feature_s`, `feature_p`
- Negative IC is equally useful — the model will learn to flip the sign automatically.

### Group 3 — horizon-dependent (color flips across horizons)
`feature_ao`, `feature_ap`, `feature_bc`, `feature_ba`, `feature_bb`, `feature_bp`, `feature_r`
- Signal reverses between short and long horizons (e.g. `feature_ao` is red at h=1, blue at h=10/25).
- In a pooled model, these signals cancel out and contribute nothing.
- Either exclude from a pooled model, or add explicit `feature × horizon` interaction terms.

### Key modeling implication
Most features have stable sign across horizons, which means a single pooled model with `horizon`
as a feature is viable. However, the horizon-flipping group is a silent trap: their positive and
negative ICs cancel in pooled training, making them appear uninformative even though they carry
real short-term or long-term signal independently.

## Panel balance check

### ts_index coverage per code
- The panel is **unbalanced**: the 4 shortest-history codes (`X9BZ68VQ`, `QAQDDTPJ`, `WH61ASEA`,

- The full time series runs from ts_index=1 to ~3,600, so the short-history codes entered the panel
  partway through — they are **newer entities**, not missing data issues.
- This matters for training: these codes have less historical signal to learn from, and a time-based
  train/val split will give them proportionally less training data than longer-history codes.

### Number of unique codes per ts_index (bottom chart)
- The dominant color is **red (h=25)**, meaning the chart is almost entirely driven by the h=25 horizon —
  the other horizons (h=1, h=3, h=10) are nearly invisible because they overlap with h=25 identically.
  This confirms all 4 horizons share the same set of codes at each time step.
- At the start of the series (ts_index < ~500), only ~10–15 codes are present, rising gradually to ~18–22
  by ts_index ~800 and staying roughly stable thereafter.
- The frequent **sharp drops to 5–10 codes** throughout the series represent time steps where many codes
  have no observation — this is normal for financial panels where not all instruments trade at every step.
- After ts_index ~2000, coverage stabilizes at 20–22 codes with sparser but still persistent dips.

### Key implications for validation
- A time-based split (e.g. last 20% of ts_index as validation) is valid, but the 4 short-history codes
  will be **absent from early training folds entirely** — their behavior may be underfit.
- The dips in code coverage mean some ts_index steps have very few rows; these steps will have
  disproportionately low representation in any grouped evaluation metric.
- Do **not** use random splits — a (code, ts_index, horizon) triplet is a single forecast instance, and
  random splitting would trivially leak same-entity same-time information across train and validation.

## Missingness co-occurrence matrix

Each cell shows the fraction of rows where **both** the row-feature and the column-feature
are missing simultaneously. A dark cell means those two features tend to be missing together.
A white cell means their missingness is independent.

### Key findings

**Cluster 1 — `feature_h / feature_i / feature_j / feature_k / feature_l`**
- Strong co-occurrence in the top-left corner (~0.12 joint missing rate).
- These 5 features are missing together on the same rows, suggesting they come from a single
  upstream data source that is sometimes entirely unavailable.
- Treat as a single missingness group: one shared binary indicator is sufficient.

**Cluster 2 — `feature_at / feature_by / feature_ay / feature_cd / feature_ce / feature_bz`**
- A visible block of co-occurrence in the middle-right region (~0.02–0.08).
- These are the high-missingness features already identified (at=12.5%, by=11%, ay=8.5%),
  and they are missing together more often than by chance.
- Again likely from a shared data source. One missingness indicator per cluster is enough,
  rather than one per feature.

**Cluster 3 — `feature_ce / feature_cf`**
- Strong co-occurrence in the bottom-right corner (~0.04).
- These two are a tightly coupled pair — if one is missing, the other usually is too.

**The vast majority of the matrix is white**
- Most features are completely independent in their missingness — no shared structure.
- This means random or isolated missingness dominates, not systematic data outages.

### Modeling implications
- Do **not** create 30+ separate missing indicators — most missingness is uncorrelated and
  adding redundant indicators adds noise without signal.
- Create **3 shared binary indicators**: one for the `feature_h/i/j/k/l` group, one for the
  `feature_at/by/ay/cd/ce/bz` group, and one for the `feature_ce/cf` pair.
- Within each group, a single median imputation strategy is appropriate since missing rows
  share the same cause.

These plots show the **marginal distribution of each top-IC feature**, overlaid across all 4 horizons.
The key question is: does the feature mean the same thing at each horizon, or does its distribution
shift? If all 4 horizon colors perfectly overlap, the feature is horizon-invariant.

### Main finding: all features are horizon-invariant in distribution
- Across all 10 features, the 4 horizon overlays are **indistinguishable** — the distributions stack
  perfectly on top of each other.
- This confirms that the same feature values appear regardless of which horizon a row belongs to,
  which makes sense: features are recorded at `ts_index` (observation time) and are shared across
  all 4 horizon rows of the same entity at the same time step.
- There is **no horizon-specific feature shift** to worry about.

### Individual feature shapes

| Feature | Shape | Notable |
|---|---|---|
| `feature_bz`, `feature_ca`, `feature_cc`, `feature_cd`, `feature_cb` | Heavy left tail, mass near 0 | Negative-skewed, mostly zero or just below |
| `feature_by` | Heavy left tail, range ~[-1.5, 0] | Bounded above at 0, long left tail |
| `feature_am`, `feature_u`, `feature_ao`, `feature_az` | Heavy right tail, mass near 0 | Positive-skewed, mostly zero or just above |

- Features with mass near zero and heavy tails are **zero-inflated** — they contain no information
  for most rows, with signal concentrated in a small fraction of extreme values.
- Log transformation will not work directly on negative-valued features (`feature_bz`, `feature_ca`,
  `feature_cc`, `feature_cd`, `feature_cb`); consider signed log: `sign(x) * log1p(|x|)`.
- For right-tailed positives (`feature_am`, `feature_u`, `feature_ao`, `feature_az`), standard
  `log1p` transform is appropriate after confirming non-negativity.

### Modeling implication
Since features are shared across horizons (same value at h=1, h=3, h=10, h=25 for the same entity
and time step), horizon-specific feature normalization is unnecessary. However, the zero-inflation
and heavy tails mean **winsorization or signed-log transforms are still needed** before feeding
these into linear models or distance-based methods.
